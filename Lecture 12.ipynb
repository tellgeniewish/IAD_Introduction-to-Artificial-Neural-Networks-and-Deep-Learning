### **Chapter 01. 랜덤 포레스트**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

wine = pd.read_csv("wine_csv_data.csv")
# wine

x = wine[["alcohol", "sugar", "pH"]].to_numpy()
y = wine["class"].to_numpy()

print(x)
```

```python
# test_size
# train_size = 1 - test_size
# 기본값 --> 학습용 : 시험용 = 75% : 25%

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2,
                                                    random_state=42)

print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier # 트리는 보통 분류, 회귀 모두 적용 ㄱㄴ

dt = DecisionTreeClassifier(random_state=42)

dt.fit(x_train, y_train)

# 과대적합(Overfitting)에 빠졌음!!
print("학습용", dt.score(x_train, y_train)) # 정확도(Accuracy)
print("시험용", dt.score(x_test, y_test))
```

```python
# 특성 중요도 --> 분류 문제를 풀 때, 각 특성들의 중요도/영향력
print(dt.feature_importances_)
# alcohol, sugar, pH

# 특성 공학 -> 과대적합 -> 1) 규제(L1, L2), 2) 특성 선택

# Feature Selection (특성 선택)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure()
plot_tree(dt)
plt.show()
```

```python
# 교차 검증(Cross Validation)
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(#n_estimators=500, # the number of estimator -> 기본값으로 100개의 나무를 만든다
                            n_jobs=-1, # n_jobs 기본값 = 1 -> 오래걸린다
                            random_state=42)

# 기본적으로 5-fold로 동작한다
scores = cross_validate(rf, x_train, y_train,
                        n_jobs=-1,
                        return_train_score=True)
```

```python
print(scores)
'''
{'fit_time': array([1.00336146, 1.24868345, 1.222754  , 1.2237494 , 1.02725267]), 
 'score_time': array([0.15159392, 0.42386651, 0.1685493 , 0.1226716 , 0.14262009]), 
 'test_score': array([0.88461538, 0.88942308, 0.90279115, 0.88931665, 0.88642926]), 
 'train_score': array([0.9971133 , 0.99663219, 0.9978355 , 0.9973545 , 0.9978355 ])}
 '''
```

```python
# Decision Tree
#학습용 0.996921300750433
#시험용 0.8584615384615385

print("학습용:", np.mean(scores['train_score']))
print("시험용:", np.mean(scores['test_score'])) # 사실은 검증용
```

```python
# Decision Tree
# [0.23101011 0.52164748 0.24734241]

rf.fit(x_train, y_train)
print(rf.feature_importances_)
```

```python
rf = RandomForestClassifier(#n_estimators=500, # the number of estimator -> 기본값으로 100개의 나무를 만든다
                            oob_score=True,
                            n_jobs=-1, # n_jobs 기본값 = 1 -> 오래걸린다
                            random_state=42)

rf.fit(x_train, y_train)

print(rf.oob_score_)
```

### **Chapter 04. 히스토그램 기반 그래디언트 부스팅**

```python
# 1) scikit-learn
# 2) Extreme Gradient Boosting -> XGBoost
# 3) Light GBM (Gradient Boosting Machine) <-- MS
# 4) CatBoost

from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)

scores = cross_validate(hgb, x_train, y_train, # n_jobs 멀티코어 사용하지 않음(매개변수 없다)!
                        return_train_score=True)
```

```python
'''
Decision Tree
학습용 0.996921300750433
시험용 0.8584615384615385

Random Forest
학습용: 0.9973541965122431
시험용: 0.8905151032797809

Extra Trees
학습용: 0.9974503966084433
시험용: 0.8887848893166506

GradientBoosting 100
학습용: 0.8881086892152563
시험용: 0.8720430147331015

GradientBoosting 500 learning_rate=0.2
학습용: 0.9464595437171814
시험용: 0.8780082549788999
'''

print("학습용:", np.mean(scores['train_score']))
print("시험용:", np.mean(scores['test_score']))
```

```python
'''
Decision Tree
[0.23101011 0.52164748 0.24734241]

Random Forest
[0.23167441 0.50039841 0.26792718]

Extra Trees
[0.20183568 0.52242907 0.27573525]

GradientBoosting 100
[0.11949946 0.74871836 0.13178218]

GradientBoosting 500 learning_rate=0.2
[0.15872278 0.68011572 0.16116151]
'''

# hgb에는 feature_importances_가 없다!

from sklearn.inspection import permutation_importance
# permutation -> n. 순열
# nPm
# nCm

# .feature_importances_을 제공하지 않는 분류, 회귀에 대해서
# permutation_importance()을 사용하면 각각의 특성 중요도 계산할 수 있다

hgb.fit(x_train, y_train)

result = permutation_importance(hgb, x_train, y_train, 
                             n_repeats=10, # 몇 번 계산할지
                             random_state=42,
                             n_jobs=-1)
print(result)                        
```

```python
print(result['importances_mean'])
# 알코올, 당도, pH
```
